```python
nimport numpy as np
import matplotlib.pyplot as plt
from keras.datasets import mnist
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
from keras.utils import np_utils

#Lấy dữ liệu từ mnist
(X_train, y_train), (X_test, y_test) = mnist.load_data()
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1] * X_train.shape[2])
X_train = X_train[:10000]

kmeans = KMeans(n_clusters=10, random_state=0)
kmeans.fit(X_train)

# Visualize the cluster centers
fig, axes = plt.subplots(2, 5, figsize=(8, 3))
for i, ax in enumerate(axes.flat):
    ax.imshow(kmeans.cluster_centers_[i].reshape(28, 28), cmap='binary', interpolation='nearest')
    ax.set_xlabel("Cluster {}".format(i))
    ax.set_xticks([])
    ax.set_yticks([])
plt.show()
```

```python
labels = kmeans.labels_

# Get indices of data points in cluster 1
cluster_1_indices = np.where(labels == 9)[0]

# Plot first 5 images in cluster 1
fig, axes = plt.subplots(1, 10, figsize=(8, 3))
for i, ax in enumerate(axes.flat):
    ax.imshow(X_train[cluster_1_indices[i]].reshape(28, 28), cmap='binary', interpolation='nearest')
    ax.set_xlabel("Image {}".format(i))
    ax.set_xticks([])
    ax.set_yticks([])
plt.show()
```
